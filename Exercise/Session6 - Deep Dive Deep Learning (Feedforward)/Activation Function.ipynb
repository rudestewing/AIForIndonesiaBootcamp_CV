```python
import numpy as np
```

## Binary Step Function

```python
def binary_step(x):

    if x<0:
        return 0
    else:
        return 1
```

```python
binary_step(5), binary_step(-10)
```

## Linear Function

```python
def linear_function(x):

    return 2*x
```

```python
linear_function(5), linear_function(-10)
```

## Sigmoid Function

```python
def sigmoid_function(x):

    z = (1/(1 + np.exp(-x)))

    return z
```

```python
sigmoid_function(5), sigmoid_function(-10)
```

## TanH Function

```python
def tanh_function(x):

    z = (2/(1 + np.exp(-2*x))) -1

    return z
```

```python
tanh_function(5), tanh_function(-10)
```

## ReLU Function

```python
def relu_function(x):

    if x<0:
        return 0
    else:
        return x
```

```python
relu_function(5), relu_function(-10)
```

## Leaky ReLU Function

```python
def leaky_relu_function(x):

    if x<0:
        return 0.01*x
    else:
        return x
```

```python
leaky_relu_function(5), leaky_relu_function(-10)
```

## Softmax Function

```python
def softmax_function(x):

    z = np.exp(x)
    logits = z/z.sum()

    return logits
```

```python
softmax_function([0.8, 1.2, 3.1])
```

## eLU Function

```python
def elu_function(x, a):

    if x<0:
        return a*(np.exp(x)-1)
    else:
        return x
```

```python
elu_function(5, 0.1), elu_function(-10, 0.1)
```

## Swish Function

```python
def swish_function(x):

    return x/(1-np.exp(-x))
```

```python
swish_function(5), swish_function(-10)
```
